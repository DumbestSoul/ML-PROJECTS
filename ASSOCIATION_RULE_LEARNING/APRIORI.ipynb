# Data pre-processing

```python
dataset = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
```

```python
dataset
```

```python
transactions = []
for i in range(0, 7501):
    transactions.append([str(dataset.values[i, j]) for j in range(0, 20)])
```

# TRAINING THE APRIORI MODEL ON THE DATASET

```python
from apyori import apriori #importing the apriori function
# this apriori model will actually return the rules from the dataset, i.e, support, confidence, lift
# parameters in the fucntion are <name of the dataset>, 
# min_lift < 3 aren't good for calcs
rules = apriori(transactions = transactions, 
                min_support = 0.003, 
                min_confidence = 0.2, 
                min_lift= 3,
                min_length = 2, 
                max_length = 2)
```

# VISUALIZING THE RESULTS

DISPLAYING THE FIRST RESULTS COMING DIRECTLY FROM THE OUTPUT OF THE APRIORI FUNCTION

```python
results = list(rules)
results
```
