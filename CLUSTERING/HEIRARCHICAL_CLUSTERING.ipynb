# IMPORTING THE LIBRARIES

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# IMPORTING THE DATASET

```python
dataset = pd.read_csv('Mall_Customers.csv')
x = dataset.iloc[:, [3, 4]].values
#visualizing the dataset plot points
plt.scatter(x[:, 0], x[:, 1], color='green', label='CUSTOMERS')
plt.xlabel('SALARIES')
plt.ylabel('SPENDING SCORE')
plt.legend()
plt.show()
```

# USING DENDROGRAM TO FIND THE OPTIMAL CLUSTER NUMBER

```python
from scipy.cluster import hierarchy as sch #hieracrchy is a function
dendrogram = sch.dendrogram(sch.linkage(x, method='ward'))
plt.title('Dendogram')
plt.xlabel('CUSTOMERS')
plt.ylabel('DISTANCE')
plt.show()
```

# TRAINING THE HIERARCHICAL CLUSTERING MODEL ON THE DATASET

```python
from sklearn.cluster import AgglomerativeClustering
aglc = AgglomerativeClustering(n_clusters = 3, affinity='euclidean', linkage='ward') #ward is the minimum variance method
y = aglc.fit_predict(x)
print(y)
```

# VISALIZING THE CLUSTERS

```python
plt.scatter(x[y==0, 0], x[y==0, 1], s=50, color='red', label='cluster1')
plt.scatter(x[y==1, 0], x[y==1, 1], s=50, color='blue', label='cluster2')
plt.scatter(x[y==2, 0], x[y==2, 1], s=50, color='green', label='cluster3')
plt.scatter(x[y==3, 0], x[y==3, 1], s=50, color='violet', label='cluster4')
plt.scatter(x[y==4, 0], x[y==4, 1], s=50, color='yellow', label='cluster5')
# plt.scatter(kmeans2.cluster_centers_[:, 0], kmeans2.cluster_centers_[:, 1],
#             s=10, color='black', label='centroids')
plt.title('K-MEANS CLUSTERING VISUALIZATION')
plt.xlabel('ANNUAL INCOME')
plt.ylabel('SPENDING THE SCORE')
plt.legend()
plt.show()
```
